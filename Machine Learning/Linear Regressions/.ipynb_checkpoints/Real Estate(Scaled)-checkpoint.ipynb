```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import f_regression
```

```python
raw_data = pd.read_csv('real_estate_price_size_year_view.csv')
```

```python
data = raw_data.copy()
data.head()
```

```python
data_mapped = data.copy()
data_mapped['view'] = data['view'].map({'No sea view':0,'Sea view':1})
data_mapped
```

```python
data.isnull().sum()
```

```python
x = data_mapped[['size','year','view']]
y = data_mapped['price']
```

```python
Scaler = StandardScaler()
```

```python
Scaler.fit(x)
x_scaled = Scaler.transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled,y, test_size = 0.2, random_state=35, shuffle=True)
```

```python
#Regressgion
reg = LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#R-Squared
reg.score(x_train,y_train)
```

```python
f_regression(x_train,y_train)
```

```python
#P-Value 
p_value = f_regression(x_train,y_train)[1]
f_statistics = f_regression(x_train,y_train)[0]
```

```python
#Summary table
reg_summary = pd.DataFrame([['Size'],['Year'],['View']], columns = ['Features'])
reg_summary
```

```python
reg_summary['Coefficients'] = reg.coef_
reg_summary
```

```python
reg_summary['P-Value'] = p_value.round(3)
reg_summary
```

```python
reg_summary['F-statistic'] = f_statistics
reg_summary
```

```python
reg_summary['R-squared'] = [reg.score(x_train,y_train),0,0]
reg_summary
```

```python
reg_summary['Intercept'] = [reg.intercept_,0,0]
reg_summary
```

```python
#Using the inputs used to create the model
train_data = x_train.copy()
train_data = pd.DataFrame(train_data)
train_data.head()
```

```python
predictions = reg.predict(train_data)
predictions
```

```python
train_data['Predictions'] = predictions
```

```python
train_data
```

```python
#Predicting with the test data
test_data = pd.DataFrame(x_test)
test_data
```

```python
test_pred = reg.predict(test_data)
test_pred = pd.DataFrame(test_pred)
test_pred
```

```python
test_data['Predictions'] = test_pred
test_data
```

```python
yhat_test = test_pred.copy()
yhat_train = predictions.copy()
```

```python
plt.scatter(y_train, yhat_train)
plt.xlabel('Y Train Data', size =18)
plt.ylabel('Predictions with Train Data', size = 18)
plt.show()
```
