```python
#Importing the relevant libraries
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
plt.rcParams['figure.figsize'] = (15,5)
from sklearn import preprocessing
```

```python
#loading data
raw_data = pd.read_csv('Data/iris-dataset.csv')
```

```python
raw_data.head()
```

```python
x = raw_data[['sepal_length','petal_length']]
y = raw_data[['sepal_width','petal_width']]
```

```python
plt.scatter(x['sepal_length'],x['petal_length'])
plt.xlabel('Sepal Length', size=18)
plt.ylabel('petal Lenght', size=18)
plt.show()
```

```python
plt.scatter(y['sepal_width'],y['petal_width'])
plt.xlabel('Sepal Width', size=18)
plt.ylabel('petal Width', size=18)
plt.show()
```

```python
plt.scatter(raw_data['sepal_length'],raw_data['sepal_width'])
plt.xlabel('Sepal Length', size=18)
plt.ylabel('petal Width', size=18)
plt.show()
```

```python
plt.scatter(raw_data['petal_length'],raw_data['petal_width'])
plt.xlabel('petal Length', size=18)
plt.ylabel('petal Width', size=18)
plt.show()
```

```python
x.head()
```

```python
x_scaled = preprocessing.scale(x)
```

```python
wcss = []
for i in range(1,10):
    kmean = KMeans(i)
    kmean.fit(x_scaled)
    wcss.append(kmean.inertia_)
```

```python
plt.plot(range(1,10),wcss)
plt.xlabel('Number of Cluster', size=18)
plt.ylabel('WCSS')
plt.show()
```

```python
kmean = KMeans(5)
kmean.fit(x_scaled)
datax_with_cluster = x.copy()
datax_with_cluster['Clusters'] = kmean.fit_predict(x_scaled)
```

```python
plt.scatter(datax_with_cluster['sepal_length'],datax_with_cluster['petal_length'], c=datax_with_cluster['Clusters'], cmap='rainbow')
plt.xlabel('Sepal Length', size=18)
plt.ylabel('Petal Length', size=18)
plt.show()
```
