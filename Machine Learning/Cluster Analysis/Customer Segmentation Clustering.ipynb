```python
#importing relevant libraries
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
plt.rcParams['figure.figsize'] = (15,5)
```

```python
#Loading Data
raw_data = pd.read_csv('Data/3.12. Example.csv')
```

```python
raw_data.head()
x = raw_data.copy()
```

```python
plt.scatter(raw_data['Satisfaction'], raw_data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

#### Number of clusters

```python
wcss = []

for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10), wcss)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

#### Clustering

```python
kmeans = KMeans(4)
kmeans.fit(x)
```

```python
identified_cluster = kmeans.fit_predict(x)
```

```python
data_with_clusters = raw_data.copy()
data_with_clusters['Clusters'] = identified_cluster
data_with_clusters.head()
```

```python
plt.scatter(data_with_clusters['Satisfaction'],data_with_clusters['Loyalty'], c = data_with_clusters['Clusters'], cmap='rainbow')
plt.xlabel('Satifaction')
plt.ylabel('Loyalty')
plt.show()
```
